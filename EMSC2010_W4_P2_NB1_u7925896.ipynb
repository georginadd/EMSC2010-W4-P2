## Plotting the Keeling curve
Monthly mean carbon dioxide has been measured at Mauna Loa Observatory, Hawaii, since 1958.  The carbon dioxide data on Mauna Loa constitute the longest record of direct measurements of CO$_2$ in the atmosphere.

This record is referred to as the *Keeling curve* in honour of Charles Keeling who started the monitoring program and ran it until his death in 2005.

```python
import pandas as pd #to be able to read the excel file
import matplotlib.pyplot as plt #for data plotting
```

```python
spreadsheet_name = 'maunaloaco2data.xlsx' # read in the excel sheet
#data obtained from NOAA (https://gml.noaa.gov/data/dataset.php?item=mlo-co2-observatory-monthly)
#[Keeling, 1976]
# Dr. Xin Lan, NOAA/GML (gml.noaa.gov/ccgg/trends/) and Dr. Ralph Keeling, Scripps Institution of Oceanography (scrippsco2.ucsd.edu/).
df = pd.read_excel(spreadsheet_name) #read the data into a dataframe
```

```python
df.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 4)) #make axes (called ax) 10 units wide and 4 units high
ax.plot(df['decimal date'],df['average'], 'r') #plot time vs. co2 average on ax as a red ('r') line
ax.set_xlabel('Time (years)') #text for the x-axis label
ax.set_ylabel('$CO_2$ monthly average (ppm)') #text for the y-axis label
ax.minorticks_on() #show the minor ticks
ax.set_title('Mauna Loa monthly $CO_2$ data')
```
